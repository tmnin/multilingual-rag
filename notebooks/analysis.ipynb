```python
results = [
    {"noise": 0.00, "cohere_r1": 1.000, "cohere_r5": 1.000, "baseline_r1": 1.000, "baseline_r5": 1.000},
    {"noise": 0.05, "cohere_r1": 0.920, "cohere_r5": 0.990, "baseline_r1": 0.690, "baseline_r5": 0.910},
    {"noise": 0.10, "cohere_r1": 0.880, "cohere_r5": 0.980, "baseline_r1": 0.480, "baseline_r5": 0.730},
    {"noise": 0.15, "cohere_r1": 0.860, "cohere_r5": 0.930, "baseline_r1": 0.240, "baseline_r5": 0.530},
    {"noise": 0.20, "cohere_r1": 0.770, "cohere_r5": 0.910, "baseline_r1": 0.200, "baseline_r5": 0.440},
]
```

```python
import matplotlib.pyplot as plt

noise = [r["noise"] for r in results]

plt.figure(figsize=(7,5))
plt.plot(noise, [r["cohere_r1"] for r in results], marker="o", label="Cohere R@1")
plt.plot(noise, [r["baseline_r1"] for r in results], marker="o", label="Baseline R@1")

plt.xlabel("Noise Probability")
plt.ylabel("Recall@1")
plt.title("Retrieval Robustness under Textual Noise (Recall@1)")
plt.legend()
plt.grid(True)
plt.show()
```
